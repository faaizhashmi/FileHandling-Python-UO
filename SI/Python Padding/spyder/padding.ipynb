```python
## Cropping tree


from PyPDF2 import PdfFileReader,PdfFileWriter
# import PyPdf

def extract_tree(in_file, out_file):
    with open(in_file, 'rb') as infp:
        # Read the document that contains the tree (in its first page)
        reader = PdfFileReader(infp)
        page = reader.getPage(0)

        # Crop the tree. Coordinates below are only referential
        page.cropBox.lowerLeft = [100,200]
        page.cropBox.upperRight = [250,300]

        # Create an empty document and add a single page containing only the cropped page
        writer = PdfFileWriter()
        writer.addPage(page)
        with open(out_file, 'wb') as outfp:
            writer.write(outfp)
           
        
def art(in_file, out_file):
    with open(in_file, 'rb') as infp:
        # Read the document that contains the tree (in its first page)
        reader = PdfFileReader(infp)
        page = reader.getPage(0)

        # Crop the tree. Coordinates below are only referential
        page.artBox.lowerLeft = [100,200]
        page.artBox.upperRight = [250,300]

        # Create an empty document and add a single page containing only the cropped page
        writer = PdfFileWriter()
        writer.addPage(page)
        with open(out_file, 'wb') as outfp:
            writer.write(outfp)
            
            
def bleed(in_file, out_file):
    with open(in_file, 'rb') as infp:
        # Read the document that contains the tree (in its first page)
        reader = PdfFileReader(infp)
        page = reader.getPage(0)

        # Crop the tree. Coordinates below are only referential
        page.bleedBox.lowerLeft = [100,200]
        page.bleedBox.upperRight = [250,300]

        # Create an empty document and add a single page containing only the cropped page
        writer = PdfFileWriter()
        writer.addPage(page)
        with open(out_file, 'wb') as outfp:
            writer.write(outfp)
            
            
def media(in_file, out_file):
    with open(in_file, 'rb') as infp:
        # Read the document that contains the tree (in its first page)
        reader = PdfFileReader(infp)
        page = reader.getPage(0)

        # Crop the tree. Coordinates below are only referential
        page.mediaBox.lowerLeft = [100,200]
        page.mediaBox.upperRight = [250,300]

        # Create an empty document and add a single page containing only the cropped page
        writer = PdfFileWriter()
        writer.addPage(page)
        with open(out_file, 'wb') as outfp:
            writer.write(outfp)

            
def allbox(in_file, out_file):
    with open(in_file, 'rb') as infp:
        # Read the document that contains the tree (in its first page)
        reader = PdfFileReader(infp)
        page = reader.getPage(0)

        # Crop the tree. Coordinates below are only referential
        page.artBox.lowerLeft = [100,200]
        page.artBox.upperRight = [250,300]
        page.bleedBox.lowerLeft = [100,200]
        page.bleedBox.upperRight = [250,300]
        page.mediaBox.lowerLeft = [100,200]
        page.mediaBox.upperRight = [250,300]

        # Create an empty document and add a single page containing only the cropped page
        writer = PdfFileWriter()
        writer.addPage(page)
        with open(out_file, 'wb') as outfp:
            writer.write(outfp)
                        
            
def insert_tree_into_page(tree_document, text_document,output_file):
    # Load the first page of the document containing 'text text text text...'
    text_page = PdfFileReader(open(text_document,'rb')).getPage(0)

    # Load the previously cropped tree (cropped using 'extract_tree')
    tree_page = PdfFileReader(open(tree_document,'rb')).getPage(0)

    # Overlay the text-page and the tree-crop   
    text_page.mergeScaledTranslatedPage(page2=tree_page,scale='1.0',tx='0',ty='0')

    # Save the result into a new empty document
    output = PdfFileWriter()
    output.addPage(text_page)
    outputStream = open(output_file,'wb')
    output.write(outputStream)
    
# First, crop the tree and save it into cropped_document.pdf
# extract_tree('wink.pdf', 'cropped_document.pdf')
art('wink.pdf', 'art.pdf')
bleed('wink.pdf', 'bleed.pdf')
media('wink.pdf', 'media.pdf')
allbox('wink.pdf', 'allbox.pdf')

# Now merge document2.pdf with cropped_document.pdf
# insert_tree_into_page('cropped_document.pdf', '2.pdf','merged.pdf')
insert_tree_into_page('art.pdf', '2.pdf','art_merged.pdf')
insert_tree_into_page('bleed.pdf', '2.pdf','bleed_merged.pdf')
insert_tree_into_page('media.pdf', '2.pdf','media_merged.pdf') 
insert_tree_into_page('allbox.pdf', '2.pdf','all_merged.pdf')    
```
